```python
# searching / transformation 
a = [ 'M', 'B', 'M', 'B'] # radius, perim... diag-> M, B
# ML-> need to convert this to numbers!-> y = mx + c
# any format ---> conv into numbers and then plot on a graph
# labels into numbers

conv = lambda val : 1 if val=='M' else 0 

```

```python
a = [ 'M', 'B', 'M', 'B']
b = []
for i in a:
  b.append(conv(i))
print(b)
```

```python
a = [ 'M', 'B', 'M', 'B']
b = [ conv(i) for i in a]
print(b)
```

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/cylons/master/data.csv', header=0)
```

```python
data.head()
```

```python
# mean, median, mode, std
```

```python
data.describe()

# z-score-> (x-mean)/standard_deviation
# 
```

```python
#[ [ 1 2 3],
# [ 2 3 4],
# [ 3 4 5] ] 

# Pixel-> R,G,B-> (100,50,200)
# one row of image-> [ p1, p2 , p3 , p4, p5....]
# image = [[ p1, p2 , p3 , p4, p5....], [ p1, p2 , p3 , p4, p5....], [ p1, p2 , p3 , p4, p5....]]...]
# Transpose-> 90 degree rotation
```

```python
data.describe().T
```

```python
stats = data.describe().T
mu = stats['mean']
sigma = stats['std']
```

```python
mu
```

```python
# z-score -> (x-mu)/sigma -> (x-mean)/standard_dev
```

```python
zscore = (data-mu)/sigma
# z-score-> -inf, inf-> where mean = 0, and every other value is just ratio wrt mean 
# 2 ways: 1) outlier removal-> any ratio>3, < -3 => outlier
# age(0-100), money(0-10million)---> normalize---> age(0,1), money(0,1); age(-3,3), money(-3,
#-3)
#normal dist (normalized probab. dist-> 0-1, -1,1)
```
